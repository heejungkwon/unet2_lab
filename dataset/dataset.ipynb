```python
import torch
from torch.utils.data import Dataset,DataLoader
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def get_path(root_dir):
  arr=[]
 
  for image in os.listdir(root_dir):
    arr.append(os.path.join(root_dir,image))
    return arr

class CustomDataset(torch.utils.data.Dataset): 
  def __init__(self,root_dir):
    self.root_dir = root_dir
    self.images = get_path(self.root_dir)
  def __len__(self):
     return len(self.images)
  def __getitem__(self, idx): 
    img = []
    for image in self.images[idx]:
      img.append(Image.open(image))

      return 
   
```

```python
#불러오기
dataset= CustomDataset('./data')
dataset.images
dataset_loader=DataLoader(dataset,batch_size=2)
```
